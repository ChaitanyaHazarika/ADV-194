# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('framingham.csv')
df

#read the csv
```

```python
#Filter and make a new dataframe for non smokers
non_smokers= df.loc[df['currentSmoker'] == 0]
non_smokers 
```

```python
#Group by age column and find average heart rate at different age

grpnonsmokers= non_smokers.groupby("age")['heartRate'].mean().reset_index()
grpnonsmokers 
```

```python
#Filter and make a new dataframe for smokers
smokers= df.loc[df['currentSmoker'] == 1]
smokers
```

```python
#Group by age column and find average heart rate at different age
grpbysmokers= smokers.groupby('age')['heartRate'].mean().reset_index()
grpbysmokers
```

```python
#Plot a line graph to show the heart rate of smokers vs non smokers
plt.figure(figsize= (20,8))
plt.plot(grpnonsmokers['age'], grpnonsmokers['heartRate'], label= "Non Smokers", linewidth= 3)
plt.plot(grpbysmokers['age'], grpbysmokers['heartRate'], label="smokers", linewidth= 3)

plt.xlabel("Age")
plt.ylabel("Heart Rate")
plt.xticks(rotation= 'vertical')
plt.title("the heart rate of smokers vs non smokers")
plt.legend()
plt.show()
```

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv

#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years
tenyearchd= df.loc[df['TenYearCHD'] == 1]
tenyearchd
```

```python
#Group by age column and count the rows of TenYearCHD column
grpbyage= tenyearchd.groupby('age')['TenYearCHD'].count().reset_index()
grpbyage
```

```python
#Plot a line graph to show total number of people having a chance of coronary heart disease in 10 years
plt.figure(figsize=(20,8))
plt.plot(grpbyage['age'], grpbyage['TenYearCHD'], label= "Colonary Heart Disease in 10 years", linewidth=3 )

plt.xlabel('Age')
plt.ylabel('Number of colonary heart disease')

plt.title('total number of people having a chance of coronary heart disease in 10 years')
plt.legend()
plt.show()
```
